```r
library(readr)
library(readxl)
library(MASS)
```

```r
data <- read_excel('/home/taz/Documentos/git/Macro/Inflacion y desempleo en francia.xlsx')
```

```r
data
```

```r
regresion_f <- lm(inflacion ~ femenino_desempleo, data = data) #Regresión
summary(regresion_f)
regresion_f$residual
residuos_f <- studres(regresion_f)
as.data.frame(residuos_f)
```

```r
plot(data$femenino_desempleo, data$inflacion, xlab='Desempleo Femenino',ylab='Inflación',main='Desempleo Femenino vs Inflación deflactor del pib')
abline(regresion_f)
```

```r
regresion <- lm(inflacion ~ masculino_desempleo, data = data) #Regresión
summary(regresion)
regresion$residual

residuos <- studres(regresion)
as.data.frame(residuos)
```

```r
plot(data$masculino_desempleo, data$inflacion, xlab='Desempleo Masculino',ylab='Inflación',main='Desempleo Masculino vs Inflación deflactor del pib')
abline(regresion)
points(5.99,12.0947650,col="red",pch=19)
```

```r
regresion_t <- lm(inflacion ~ desempleo_total, data = data) #Regresión
summary(regresion_t)
regresion_t$residual
residuos_t <- studres(regresion_t)
as.data.frame(residuos_t)
```

```r
plot(data$desempleo_total, data$inflacion, xlab='Desempleo Total',ylab='Inflación',main='Desempleo Total vs Inflación deflactor del pib')
abline(regresion_t)
points(7.54,11.6919725,col="red",pch=19)
points(8.2,12.094765,col="red",pch=19)
```

```r
matplot(data$Año, cbind(data$femenino_desempleo,data$inflacion),type="l",col=c("red","green"),lty=c(1,1),xlab='Años',ylab='Indicadores',main='Desempleo Femino e Inflación',las=1)
legend(x = 'topright', legend = c("desempleo Femenino", "Inflación"), fill = c("red", "green"))

matplot(data$Año, cbind(data$masculino_desempleo,data$inflacion),type="l",col=c("red","green"),lty=c(1,1),xlab='Años',ylab='Indicadores',main='Desempleo Masculino e Inflación',las=1)
legend(x = 'topright', legend = c("desempleo Masculino", "Inflación"), fill = c("red", "green"))

matplot(data$Año, cbind(data$desempleo_total,data$inflacion),type="l",col=c("red","green"),lty=c(1,1),xlab='Años',ylab='Indicadores',main='Desempleo Total e Inflación',las=1)
legend(x = 'topright', legend = c("desempleo Total", "Inflación"), fill = c("red", "green"))

matplot(data$Año, cbind(data$femenino_desempleo,data$masculino_desempleo),type="l",col=c("red","green"),lty=c(1,1),xlab='Años',ylab='Indicadores',main='Desempleo Femenino y Masculino',las=1)
legend(x = 'topright', legend = c("desempleo Femenino", "Desempleo Masculino"), fill = c("red", "green"))


```

```r
matplot(data$Año, cbind(data$femenino_desempleo,data$inflacion,data$masculino_desempleo,data$desempleo_total),type="l",col=c("red","green","blue","purple"),lty=c(1,1),xlab='Años',ylab='Indicadores',main='Desempleo Femino e Inflación',las=1)
legend(x = 'topright', legend = c("desempleo Femenino", "Inflación","desempleo Masculino","desempleo total"), fill = c("red", "green","blue","purple"))
```
